```python
# Crear columna 'data_naixement' en format datetime
# .to_datetime() reconeix automàticament un DataFrame si les seves columnes es diuen exactament:year, month, day per això les renombrem a year, month, day
enquesta['data_naixement'] = pd.to_datetime(
    enquesta[['Any', 'Mes', 'Dia']].rename(columns={'Any': 'year', 'Mes': 'month', 'Dia': 'day'}))
enquesta.head()
```

```python
# Utilitzant la columna data de naixement (en format datetime) creem una funció que donada una data, calculi l'edat actual a dia d'avui.

def calcular_edat(data_naix): # Funció per calcular l'edat a partir de la data de naixement
    avui = pd.Timestamp.today()  # data actual (és un objecte datetime de pandas amb la data i hora actual)
    edat = avui.year - data_naix.year # calcular diferència en anys i ajustar si encara no ha fet anys aquest any
    if (avui.month, avui.day) < (data_naix.month, data_naix.day): #si el mes i dia actual és abans del mes i dia de naixement restem un any
        edat -= 1
    return edat

# Afegir columna 'edat' al DataFrame
enquesta['edat'] = enquesta['data_naixement'].apply(calcular_edat) #aplicar la funció calcular_edat a cada fila de la columna 'data_naixement' i crea una nova columna 'edat'
enquesta.head()

```

1.Crea una funció que prengui un dataframe com a paràmetre d'entrada.

La funció ha de crear (i exportar) un gràfic automàticament per a cada columna del dataframe. Per exemple:

un histograma/boxplot si la variable és numèrica
unes barres dels valors més freqüents si és categòrica
unes barres dels anys més freqüents si la dada està en format data.
La idea és crear una funció que funcioni per qualsevol dataframe, no només amb el que hem treballat fins ara.

Mostra el resultat de la funció en algun dels datasets d’exemple que conté el paquet seaborn. Per exemple, iris, penguins o titanic.

Tingues en consideració que en el següent sprint treballaràs exclusivament amb gràfics. L’objectiu d’aquest exercici no és crear gràfics molt elaborats, sinó resoldre una necessitat de manera ràpida i automàtica.

```python
import pandas as pd
import matplotlib.pyplot as plt

def auto_plots(df): # Funció per generar gràfics automàtics segons el tipus de dada
    """
    Genera un gràfic automàtic per a cada columna del dataframe:
      - Numèriques  → histograma
      - Categòriques → barres de freqüències
      - Dates → barres per any
    """

    for col in df.columns:
        series = df[col].dropna() #Treu els valors nuls per no afectar els gràfics

        plt.figure(figsize=(8, 5))

        # Variables numèriques
        if pd.api.types.is_numeric_dtype(series): #comprovar si la sèrie és numèrica i fer l'histograma
            plt.hist(series)
            plt.title(f"Histograma de {col}")
            plt.xlabel(col)
            plt.ylabel("Freqüència")

        # Variables data
        elif pd.api.types.is_datetime64_any_dtype(series): #comprovar si la sèrie és de tipus data i fem barres per any (extraiem l’any amb .dt.year).
            years = series.dt.year.value_counts().sort_index()
            plt.bar(years.index.astype(str), years.values)
            plt.title(f"Freqüència per any de {col}")
            plt.xlabel("Any")
            plt.ylabel("Comptatge")

        # Variables categòriques / objecte
        else:
            counts = series.value_counts() # Comptar la freqüència de cada categoria
            plt.bar(counts.index.astype(str), counts.values) #crear barres de freqüència   
            plt.title(f"Barres de {col}")
            plt.xlabel(col)
            plt.ylabel("Freqüència")
            plt.xticks(rotation=45) # Rotar les etiquetes de l'eix x per millor visibilitat

        plt.tight_layout() # Ajusta els marges per evitar que els títols/etiquetes se solapin
        plt.show() # Mostra el gràfic per cada iteració
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris["target"] = iris.target #afegim la columna target per tenir una columna categòrica
auto_plots(df_iris)
```

```python
import pandas as pd
import seaborn as sns

df_penguins = sns.load_dataset("penguins") # Carregar el dataset Penguins des de seaborn

auto_plots(df_penguins)
```

Font: Mejores Rutas

Ens interessa visitar totes les ciutats principals d'Espanya recorrent la mínima distància possible. No cal que ho facis de forma òptima, ens interessa que desenvolupis una solució raonable utilitzant les eines que tens actualment. Per exemple, una aproximació senzilla (que no òptima) seria anant sempre a la ciutat més propera que no haguem visitat encara. Fes una funció que donada la matriu de distàncies i la ciutat d'origen, faci una proposta de ruta que sigui el més curta possible que puguis, retornant una llista amb l'ordre de visita. Dóna també la distància total recorreguda.

(EXTRA) Des de quina ciutat la ruta seria més curta amb l'algoritme plantejat

```python
import pandas as pd
df = pd.read_excel("matriu_distancies.xlsx", sheet_name="Distancia (kilómetros)") # Carregar l’arxiu amb noms de files i columnes
df
```

```python
df=df.rename(columns={"Unnamed: 0": "Ciutat"}) # Renombrar la columna d'índex
df
```

```python
# Ciutats que volem eliminar
ciutats_a_borrar = ["Las Palmas de Gran Canaria", "Palma"]
df = df.set_index('Ciutat') # Estableix l'índex a la columna de ciutats
df = df.reset_index() #restablir l'index en una nova columna
df = df[~df['Ciutat'].isin(ciutats_a_borrar)]# Eliminar files on la columna 'Ciutat' coincideix amb ciutats_a_borrar
df = df.drop(columns=ciutats_a_borrar, errors='ignore') # Eliminar columnes corresponents (ignora si alguna columna no existeix)
df
```

```python
import numpy as np

# Convertir la matriu de distàncies a numpy array (excloent la columna 'Ciutat')
df_dist = df.set_index('Ciutat')
dist_matrix = df_dist.values

# Definir la ciutat d'inici (per exemple, 'Barcelona')
ciutat_inici = df_dist.index.get_loc('Barcelona')

# Funció de veí més proper (nearest neighbor)
def ruta_proximitat(dist_matrix, ciutat_inici=0):
    n = dist_matrix.shape[0]
    visitades = [ciutat_inici]
    dist_total = 0
    ciutat_actual = ciutat_inici

    while len(visitades) < n:
        # Crear la lista de ciudades no visitadas usando un bucle
        no_visitades = []
        for i in range(n):
            if i not in visitades:
                no_visitades.append(i)
        
        # Buscar la ciudad más cercana
        distancies = dist_matrix[ciutat_actual, no_visitades]
        idx_min = np.argmin(distancies)
        propera_ciutat = no_visitades[idx_min]
        
        # Actualizar distancia y marcar ciudad como visitada
        dist_total += dist_matrix[ciutat_actual, propera_ciutat]
        visitades.append(propera_ciutat)
        ciutat_actual = propera_ciutat

    # Tornar al punt d'origen
    dist_total += dist_matrix[ciutat_actual, ciutat_inici]
    visitades.append(ciutat_inici)

    return visitades, dist_total

# Generar ruta
ruta_idx, dist_total = ruta_proximitat(dist_matrix, ciutat_inici)
ruta_ciutats = [df_dist.index[i] for i in ruta_idx]

print("Ruta:", ruta_ciutats)
print("Distància total:", dist_total)
```

```python
# És un patró típic d’algoritmes del Travelling Salesman Problem (TSP) amb heurística de proximitat (nearest neighbor).
# Provar totes les ciutats com a origen i quedar-se amb la millor ruta (la més curta)
# Si canvies la ciutat inicial, canvia la ruta i la distància total

millor_dist = float('inf') #inicialitzar amb infinit. Suposem que la millor distància és infinity perquè qualsevol ruta real serà més curta.
millor_ruta = None #inicialitzar amb None. Encara no sabem quina és la millor ruta.
millor_ciutat = None #inicialitzar amb None. Encara no sabem quin és el millor origen.
n = dist_matrix.shape[0] # Nombre total de ciutats.

for i in range(n):
    ruta_idx, dist_total = ruta_proximitat(dist_matrix, i)
    if dist_total < millor_dist:
        millor_dist = dist_total
        millor_ruta = ruta_idx
        millor_ciutat = i

millor_ruta_ciutats = []
for i in millor_ruta:
    ciutat = df_dist.index[i]
    millor_ruta_ciutats.append(ciutat)
# millor_ruta_ciutats = [df_dist.index[i] for i in millor_ruta]

print("Millor ciutat d'inici:", df_dist.index[millor_ciutat])
print("Distància total més curta segons l'heurística:", millor_dist)
print("Ruta completa:", millor_ruta_ciutats)
```
