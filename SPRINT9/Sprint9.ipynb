1. Calculadora de l'índex de massa corporal
Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en: IMC= pes/(alçada)2

La funció ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduïdes per l’usuari/a no estan en el format adequat o no pren valors raonables.

```python
def calcular_imc():
    try:
        pes = float(input("Entra el teu pes en kg: ").replace(",", "."))#canviem comes per punts si 
        if pes <= 0 or pes > 300:
            print("El pes introduït no és raonable.")
            return
        
        alçada = float(input("Introdueix la teva alçada en metres: ").replace(",", ".")) #canviem comes per punts si cal
        if alçada <= 0 or alçada > 2.5:
            print("L'alçada introduïda no és raonable.")
            return

    except ValueError:
        print("Entrada no vàlida. Si us plau, introdueix un número raonable.")
        return

    imc = pes / (alçada ** 2)
    print(f"El teu IMC és {imc:.2f}")

    if imc < 18.5:
        print("Tens pes baix")
    elif imc < 24.9:
        print("Tens un pes normal") 
    elif imc < 29.9:
        print("Tens sobrepès")
    else:
        print("Tens obesitat")

calcular_imc()

```

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

```python
# versió match-case 
def convertidor_temperatures():
    # Demanar temperatura dins d'un try-except
    try:
        temperatura = float(input("Introdueix la temperatura a convertir: "))
        if temperatura < -273.15 or temperatura > 10000:
            print("La temperatura introduïda no és raonable.")
            return
    except ValueError:
        print("Has d'introduir un número vàlid.")
        return

    # Demanar unitat d'origen
    unitat = input("Introdueix la unitat d'origen (C, F, K): ").strip().upper()

    # Validació amb match-case
    match unitat:
        case "C" | "F" | "K":
            pass   # és vàlida
        case _:
            print("Unitats vàlides: C, F o K.")
            return

    # Convertim a Celsius utilitzant match-case
    match unitat:
        case "C":
            temp_celsius = temperatura
        case "F":
            temp_celsius = (temperatura - 32) * 5 / 9
        case "K":
            temp_celsius = temperatura - 273.15

    # Convertir a totes les unitats
    temp_fahrenheit = (temp_celsius * 9 / 5) + 32
    temp_kelvin = temp_celsius + 273.15

    return temperatura, unitat, temp_celsius, temp_fahrenheit, temp_kelvin

resultats = convertidor_temperatures()

if resultats: 
    temperatura, unitat, temp_celsius, temp_fahrenheit, temp_kelvin = resultats

    print(f"La temperatura {temperatura} {unitat} es converteix a:")

    if unitat != "C":
        print(f"{temp_celsius:.1f} C")
    if unitat != "F":
        print(f"{temp_fahrenheit:.1f} F")
    if unitat != "K":
        print(f"{temp_kelvin:.1f} K")
```

3. Comptador de paraules d'un text.
Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
import string #proporciona una llista de tots els signes de puntuació que volem eliminar del text.

def comptar_paraules(text):
    if not text or text.strip() == "":
        print("El text està buit.")
        return
    
    text = text.lower()  # Normalitzar: tot en minúscules

    for signe in string.punctuation: # string.punctuation conté tots els signes de puntuació que volem eliminar
        text = text.replace(signe, "") #Elimina els signes de puntuació del text

    paraules = text.split() #Elimina automàticament els espais entre paraules i retorna només les paraules
    comptador = {}
    for paraula in paraules:
        if paraula in comptador:
            comptador[paraula] += 1
        else:
            comptador[paraula] = 1

    print("Comptador de paraules:")
    for paraula, vegades in comptador.items():
        print(f"La paraula {paraula}: apareix {vegades} vegades")

    #EXTRA: llargada mitjana
    llargades = []  # Cream una llista buida
    for paraula in paraules:
        llargades.append(len(paraula))  # afegim la llargada a la llista
    
    mitjana = sum(llargades) / len(paraules)
    print(f"\nLlargada mitjana de les paraules: {mitjana:.1f}")

    return comptador, mitjana

text = input("Introdueix un text per analitzar: ")
comptar_paraules(text)
```

4. Diccionari invers (amb possibilitat de duplicats)
Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.
Fixa’t en el següents exemples per veure l’output esperat.

reverse_dictionari=({'a':1, 'b':2, 'c':3})
{1:'a', 2:'b', 3:'c'}

```python
# Volem invertir un diccionari, però tenint en compte que pot haver-hi valors duplicats,perque si dos valors son iguals, i els passem a claus
# no podem tenir dues claus diferents que tenen el mateix valor, llavors quin dels dos nous valors guardem?
# no pots invertir el diccionari sense perdre informació.

def diccionari_invers(dic):
    invers = {}
    duplicats = {}

    for clau, valor in dic.items():
        if valor in invers:
            # Registrem duplicats
            if valor not in duplicats:
                duplicats[valor] = [invers[valor]]
            duplicats[valor].append(clau)
        else:
            invers[valor] = clau

    if duplicats:
        print("Advertiment: s'han trobat valors duplicats!")
        for val, claus in duplicats.items():
            print(f"Valor repetit '{val}' associat a les claus: {claus}")

    return invers

print("\nPrimer exemple amb duplicats:")
diccionari_original = {'a': 1, 'b': 2, 'c': 1, 'd': 3}
diccionari_invertit = diccionari_invers(diccionari_original)
print(f"El diccionari original és: {diccionari_original}")
print(f"El diccionari invertit és: {diccionari_invertit}")

print("\nSegon exemple sense duplicats:")

diccionari_original = {'a': 1, 'b': 2, 'c': 4, 'd': 3}
diccionari_invertit = diccionari_invers(diccionari_original)
print(f"El diccionari original és: {diccionari_original}")
print(f"El diccionari invertit és: {diccionari_invertit}")
```

1. Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
import string
from collections import defaultdict, Counter
# string → dóna accés a coses útils com:
# string.ascii_uppercase = "A B C ... Z"
# string.punctuation = tots els signes de puntuació
# string.digits = "0 1 2 ... 9"
# Counter → és com un diccionari especial que compta elements automàticament.

def comptar_paraules_per_lletra(nom_fitxer):
    diccionari = {lletra: {} for lletra in string.ascii_uppercase} # Crear un diccionari amb lletres A-Z com a claus i subdiccionaris buits
    
    with open(nom_fitxer, 'r', encoding='utf-8') as f: #obrir el fitxer en mode lectura amb codificació utf-8  
        text = f.read() #llegir tot el contingut del fitxer i emmagatzemar-lo a la variable text

    # Normalitzar el text: passar a majúscules i substituir caràcters no alfabètics per espais
    text = text.upper()
    for char in string.punctuation + string.digits:
        text = text.replace(char, ' ')
        
    # Dividir en paraules
    paraules = text.split()
    
    # Comptar la freqüència de cada paraula
    freq = Counter(paraules)

    # Assignar les paraules al subdiccionari corresponent a la seva primera lletra
    for paraula, count in freq.items():
        if paraula[0] in diccionari:
            diccionari[paraula[0]][paraula] = count

    # Ordenar les paraules dins de cada lletra
    for lletra in diccionari:
        diccionari[lletra] = dict(sorted(diccionari[lletra].items()))

    return diccionari

nom_fitxer = r"C:\Users\NURIA\Desktop\A_IT_ACADEMY\DATA ANALITICS\Sprint 9\tu me quieres blanca.txt"
resultat = comptar_paraules_per_lletra(nom_fitxer)

# Mostrar resultat
for lletra, subdic in resultat.items():
    if subdic:  # només mostrar lletres amb paraules
        print(f"{lletra}:")
        for paraula, freq in subdic.items():
            print(f"  {paraula}: {freq}")
```

2. Conversió de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

[ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python
dades = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']] 

convertibles = [] #llista per emmagatzemar els elements convertibles a float
no_convertibles = []

def process_element(element): #funció recursiva per processar elements
    if isinstance(element, (list, tuple)): #
        # Recorrem recursivament elements de llistes o tuples
        for item in element: #recorrem cada element dins de la llista o tupla
            process_element(item) #crida recursiva per processar cada element
    else:
        try:
            num = float(element) #intentem convertir l'element a float
            convertibles.append(num) #si té èxit, l'afegim a la llista de convertibles
        except (ValueError, TypeError): #si no es pot convertir, l'afegim a la llista de no convertibles
            no_convertibles.append(element) #afegim a la llista de no convertibles
#L’element és un valor que no es pot convertir → ValueError o l’element no és un valor simple, sinó un tipus incorrecte → TypeError 

for dada in dades: #recorrem cada element de la llista dades
    process_element(dada) #processar cada element

# Mostrar resultats
print("Elements convertibles:", convertibles)
print("Elements no convertibles:", no_convertibles)
```

```python
import numpy as np #s’utilitza per triar caràcters aleatoris i barrejar la llista.
import string #proporciona conjunts de caràcters com lletres majúscules, minúscules, dígits i signes de puntuació.

def crear_contrasenya(longitud, majuscules=True, minuscules=True, numeros=True, signes=False):
    char_sets = [] #llista per emmagatzemar els conjunts de caràcters seleccionats
    if majuscules:
        char_sets.append(string.ascii_uppercase)
    if minuscules:
        char_sets.append(string.ascii_lowercase)
    if numeros:
        char_sets.append(string.digits)
    if signes:
        char_sets.append(string.punctuation)

    if not char_sets:
        raise ValueError("Cal activar almenys un tipus de caràcter!") # garantir que s'ha seleccionat almenys un tipus de caràcter
  
    contrasenya = []
    for chars in char_sets:
        char_aleatori = np.random.choice(list(chars)) #triar un caràcter aleatori de cada conjunt   
        contrasenya.append(char_aleatori) #afegir el caràcter triat a la contrasenya
    # alternativa as comprehension list: contrasenya = [np.random.choice(list(chars)) for chars in char_sets]

    # Hem de generar la resta de caràcters fins assolir la longitud desitjada
    tots_els_chars = ''.join(char_sets)
    restant = longitud - len(contrasenya) #nombre de caràcters restants per afegir 

    contrasenya.extend(np.random.choice(list(tots_els_chars), restant)) #afegir caràcters aleatoris restants

    np.random.shuffle(contrasenya) #barrejar els caràcters perquè l’ordre sigui aleatori

    return ''.join(contrasenya) #retornar la contrasenya com una cadena de text

longitud = 12   
contrasenya = crear_contrasenya(longitud, majuscules=True, minuscules=True, numeros=True, signes=True)
print("Contrasenya generada:", contrasenya) 
```

2. Processament de dades simple

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. Té un històric de partits de futbol català en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera automàtica, per extreure’n els resultats que necessita.

Utilitza el fitxer "historic_partits.txt"

Necessita un programa que retorni:

El nombre total de gols que ha fet cada equip.
El nom de l’equip més golejador.
El nom de l’equip més golejat
La classificació global (cada victòria: 3 pts, empat 1 pts, derrota 0 pts)
Com a referència, els equips amb més punts serien els següents:

```python
import re #s’utilitza per treballar amb expressions regulars, que són patrons utilitzats per cercar i manipular cadenes de text.
from collections import defaultdict #

gols_fets = defaultdict(int) #diccionari per comptar els gols fets per cada equip
gols_rebuts = defaultdict(int) #diccionari per comptar els gols rebuts per cada equip
punts = defaultdict(int) #diccionari per comptar els punts de cada equip

# Expressió regular per capturar "Equip1 G1-G2 Equip2" pero hi ha equips amb espais en blanc ex "Reus Deportiu	4-4	Nàstic de Tarragona"
patro = re.compile(r"(.+?)\s+(\d+)-(\d+)\s+(.+)") # serveix per extreure els noms dels equips i els gols de cada partit
#. significa qualsevol caràcter.+ significa “un o més” caràcters.? aquí fa que sigui no-greedy, és a dir, captura la mínima quantitat de caràcters 
# possibles que satisfacin la resta de l’expressió. Propòsit: capturar el nom del primer equip, encara que tingui espais.

with open("C:\\Users\\NURIA\\Desktop\\A_IT_ACADEMY\\DATA ANALITICS\\Sprint 9\\historic partits.txt", "r", encoding="utf-8") as f:
    for linia in f:
        linia = linia.strip()
        if not linia:
            continue

        match = patro.match(linia)
        if not match:
            print("Error de format:", linia)
            continue

        equip1 = match.group(1)
        gol1 = int(match.group(2))
        gol2 = int(match.group(3))
        equip2 = match.group(4)

        # Actualitzem gols
        gols_fets[equip1] += gol1
        gols_fets[equip2] += gol2
        gols_rebuts[equip1] += gol2
        gols_rebuts[equip2] += gol1

        # Assignem punts
        if gol1 > gol2:
            punts[equip1] += 3
        elif gol2 > gol1:
            punts[equip2] += 3
        else:
            punts[equip1] += 1
            punts[equip2] += 1

# Equip més golejador
equip_mes_golejador = max(gols_fets, key=lambda k: gols_fets[k])
# Equip més golejat
equip_mes_golejat = max(gols_rebuts, key=lambda k: gols_rebuts[k])
# Classificació global ordenada
classificacio = sorted(punts.items(), key=lambda x: x[1], reverse=True)

# Resultats
print("Gols per equip:")
for equip, gols in gols_fets.items():
    print(f"{equip}: {gols} gols")

print(f"\nEquip més golejador: {equip_mes_golejador} ({gols_fets[equip_mes_golejador]} gols)")
print(f"Equip més golejat: {equip_mes_golejat} ({gols_rebuts[equip_mes_golejat]} gols rebuts)")

print("\nClassificació global:")
for i, (equip, pts) in enumerate(classificacio, start=1):
    print(f"{i}. {equip}: {pts} pts")
```
