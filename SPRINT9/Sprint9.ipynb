Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).
Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

```python
def convertidor_temperatures():
    # Demanar temperatura dins d'un try-except
    try:
        temperatura = float(input("Introdueix la temperatura a convertir: "))
        if temperatura < -273.15 or temperatura > 10000:
            print("La temperatura introduïda no és raonable.")
            return
    except ValueError:
        print("Has d'introduir un número vàlid.")
        return

    # Demanar unitat d'origen
    unitat = input("Introdueix la unitat d'origen (C, F, K): ").strip().upper()

    # Validació amb match-case
    match unitat:
        case "C" | "F" | "K":
            pass
        case _:
            print("Unitats vàlides: C, F o K.")
            return

    # Convertir a Celsius amb un diccionari de funcions
    conversio_a_celsius = { 
        "C": lambda t: t,
        "F": lambda t: (t - 32) * 5 / 9,
        "K": lambda t: t - 273.15
    }

    temp_celsius = conversio_a_celsius[unitat](temperatura)

    # Convertir a totes les unitats
    temp_fahrenheit = (temp_celsius * 9 / 5) + 32
    temp_kelvin = temp_celsius + 273.15

    # Mostrar resultats
    print(f"La temperatura {temperatura} {unitat} es converteix a:")

    if unitat != "C":
        print(f"{temp_celsius:.1f} C")
    if unitat != "F":
        print(f"{temp_fahrenheit:.1f} F")
    if unitat != "K":
        print(f"{temp_kelvin:.1f} K")

convertidor_temperatures()
```

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
def comptar_paraules(text):
    import string
    from pprint import pprint   # ➜ Afegim pprint per imprimir directament el diccionari

    if not text or text.strip() == "":
        print("El text està buit.")
        return
    
    text = text.lower()

    # Eliminar signes de puntuació
    for signe in string.punctuation:
        text = text.replace(signe, "")
    
    # Separar el text en paraules
    paraules = text.split()

    # Comptador de paraules
    comptador = {}
    for paraula in paraules:
        if paraula in comptador:
            comptador[paraula] += 1
        else:
            comptador[paraula] = 1

    print("\nComptador de paraules:")
    pprint(comptador)   # ➜ Imprimir directament el diccionari ben format

    # EXTRA: llargada mitjana
    llargades = [len(p) for p in paraules]
    mitjana = sum(llargades) / len(paraules)

    print(f"\nLlargada mitjana de les paraules: {mitjana:.1f}")

    return comptador, mitjana


def main():
    text = input("Introdueix un text per analitzar: ")
    comptar_paraules(text)
main()
```

4. Diccionari invers (amb possibilitat de duplicats)
Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.
Fixa’t en el següents exemples per veure l’output esperat.

reverse_dictionari=({'a':1, 'b':2, 'c':3})
{1:'a', 2:'b', 3:'c'}

```python
def invertir_diccionari(dic):
    invers = {}

    for clau, valor in dic.items():
        # Si el valor ja existeix a l'invers → hi ha duplicats
        if valor in invers:
            # Si el que hi ha és una llista, només afegim la clau nova
            if isinstance(invers[valor], list):
                invers[valor].append(clau)
            else:
                # Si no és una llista, convertim l’entrada en una llista
                invers[valor] = [invers[valor], clau]

            print(f"Valor duplicat trobat: {valor}")
        else:
            # Si no hi és, el posem directament
            invers[valor] = clau

    return invers


# Exemple d’ús
original = {
    "a": 1,
    "b": 2,
    "c": 1,
    "d": 3,
    "e": 2
}

dicc_invers = invertir_diccionari(original)
print("\nDiccionari invers:")
print(dicc_invers)
```

1. Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
import string
import re
from collections import Counter #és com un diccionari especial que compta elements automàticament (serveix per comptar freqüències)
from pprint import pprint


def comptar_paraules_per_lletra(nom_fitxer):
    # Diccionari A-Z amb subdiccionaris buits
    diccionari = {lletra: {} for lletra in string.ascii_uppercase}
    
    # Llegir fitxer
    with open(nom_fitxer, 'r', encoding='utf-8') as f:
        text = f.read()

    # Trobar totes les paraules en majúscules amb regex
    paraules = re.findall(r'\b\w+\b', text.upper()) # es más eficiente y limpio porque realiza la conversión a minúsculas, la limpieza de caracteres 
    # no deseados y la división en palabras en una única línea y evita la creación de mas variables.

    freq = Counter(paraules)

    # Assignar paraules al subdiccionari de la seva primera lletra
    for paraula, count in freq.items():
        inicial = paraula[0]
        if inicial in diccionari:
            diccionari[inicial][paraula] = count

    # Ordenar les paraules dins de cada lletra
    for lletra in diccionari:
        diccionari[lletra] = dict(sorted(diccionari[lletra].items()))

    return diccionari


def main():
    nom_fitxer = "tu me quieres blanca.txt"
    resultat = comptar_paraules_per_lletra(nom_fitxer)
    # Mostrar resultat amb pprint
    pprint(resultat)

main()
```

string.punctuation = tots els signes de puntuació
string.digits = "0 1 2 ... 9"

2. Conversió de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

[ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python
def es_convertible_a_float(element):
    """Intenta convertir l'element a float i retorna (True, valor) o (False, element)."""
    try:
        num = float(element)
        return True, num
    except (ValueError, TypeError):
        return False, element

def process_element(element, convertibles, no_convertibles):
    """Processa un element, sigui simple, llista o tupla."""
    if isinstance(element, (list, tuple)):
        for item in element:
            process_element(item, convertibles, no_convertibles)
    else:
        convertible, resultat = es_convertible_a_float(element)
        if convertible:
            convertibles.append(resultat)
        else:
            no_convertibles.append(resultat)


def processar_dades(dades):
    """Processa tota la llista inicial i retorna dues llistes."""
    convertibles = []
    no_convertibles = []

    for dada in dades:
        process_element(dada, convertibles, no_convertibles)

    return convertibles, no_convertibles


def main():
    dades = ['1.3', 'one', '1e10', 'seven', '3-1/2',
             ('2', 1, 1.4, 'not-a-number'),
             [1, 2, '3', '3.4']]

    convertibles, no_convertibles = processar_dades(dades)

    print("Elements convertibles:", convertibles)
    print("Elements no convertibles:", no_convertibles)
main()

```

Explora el funcionament del mòdul random de la llibreria numpy.

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

longitud (int): Longitud de la contrasenya
majuscules (bool = True): Si hi ha d’aparèixer majúscules
minuscules (bool = True): Si hi ha d’aparèixer minúscules
numeros (bool = True): Si hi ha d’aparèixer números
signes (bool = False): Si hi ha d’aparèixer caràcters especials (,-$? o similars)
Així doncs, si executem la funció de la següent manera:

crear_contrasenya(10, True, True, True, True)

Hauríem d’obtenir un output (que hem de poder guardar) de l’estil:

9Er,5Vn8P$

Assegura’t que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleatòries.

(EXTRA) Explora com podríem fer que la funció copiés la contrasenya automàticament al porta papers de l’ordinador (com si l’haguéssim seleccionat i fet ctrl+copy).

```python
import numpy as np
import string

def obtenir_conjunts_caracters(majuscules, minuscules, numeros, signes):
    """Retorna una llista amb els conjunts de caràcters seleccionats."""
    char_sets = []
    if majuscules:
        char_sets.append(string.ascii_uppercase)
    if minuscules:
        char_sets.append(string.ascii_lowercase)
    if numeros:
        char_sets.append(string.digits)
    if signes:
        char_sets.append(string.punctuation)

    if not char_sets:
        raise ValueError("Cal activar almenys un tipus de caràcter!")
    
    return char_sets


def obtenir_caracters_obligatoris(char_sets):
    """Retorna un caràcter aleatori de cada conjunt per garantir diversitat."""
    obligatoris = []
    for chars in char_sets:
        aleatori = np.random.choice(list(chars))
        obligatoris.append(aleatori)
    return obligatoris


def generar_caracters_restants(char_sets, restant):
    """Genera caràcters aleatoris fins a completar la longitud."""
    tots_els_chars = ''.join(char_sets)
    caracters = []
    for i in range(restant):
        caracters.append(np.random.choice(list(tots_els_chars)))
    return caracters

def barrejar_caracters(caracters):
    """Barreja la llista de caràcters."""
    np.random.shuffle(caracters)
    return caracters


def crear_contrasenya(longitud, majuscules=True, minuscules=True, numeros=True, signes=False):
    char_sets = obtenir_conjunts_caracters(majuscules, minuscules, numeros, signes)
    contrasenya = obtenir_caracters_obligatoris(char_sets)
    restant = longitud - len(contrasenya)
    if restant > 0:
        contrasenya.extend(generar_caracters_restants(char_sets, restant))
    contrasenya = barrejar_caracters(contrasenya)
    return ''.join(contrasenya)


# ---------- EXECUCIÓ D'EXEMPLE ----------
longitud = 12
contrasenya = crear_contrasenya(longitud, majuscules=True, minuscules=True, numeros=True, signes=True)
print("Contrasenya generada:", contrasenya)
```

2. Processament de dades simple

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. Té un històric de partits de futbol català en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera automàtica, per extreure’n els resultats que necessita.

Utilitza el fitxer "historic_partits.txt"

Necessita un programa que retorni:

El nombre total de gols que ha fet cada equip.
El nom de l’equip més golejador.
El nom de l’equip més golejat
La classificació global (cada victòria: 3 pts, empat 1 pts, derrota 0 pts)
Com a referència, els equips amb més punts serien els següents:

```python
import re #s’utilitza per treballar amb expressions regulars, que són patrons utilitzats per cercar i manipular cadenes de text.
from collections import defaultdict #from collections import defaultdict

def llegir_partits(nom_fitxer):
    """Llegeix un fitxer i retorna una llista de tuples (equip1, gol1, gol2, equip2)."""
    patro = re.compile(r"(.+?)\s+(\d+)-(\d+)\s+(.+)") # Expressió regular per capturar "Equip1 G1-G2 Equip2" pero hi ha equips amb espais en blanc ex "Reus Deportiu	4-4	Nàstic de Tarragona"
    """serveix per extreure els noms dels equips i els gols de cada partit(#. significa qualsevol caràcter.+ significa “un o més” caràcters.? aquí fa que sigui no-greedy, és a dir, captura la mínima quantitat de caràcters 
    possibles que satisfacin la resta de l’expressió. Propòsit: capturar el nom del primer equip, encara que tingui espais"""
    partits = []

    with open(nom_fitxer, "r", encoding="utf-8") as f:
        for linia in f:
            linia = linia.strip()
            if not linia:
                continue

            match = patro.match(linia)
            if not match:
                print("Error de format:", linia)
                continue

            equip1 = match.group(1)
            gol1 = int(match.group(2))
            gol2 = int(match.group(3))
            equip2 = match.group(4)

            partits.append((equip1, gol1, gol2, equip2))
    return partits

def calcular_gols(partits):
    """Calcula els gols fets i rebuts per cada equip."""
    gols_fets = defaultdict(int)
    gols_rebuts = defaultdict(int)

    for equip1, gol1, gol2, equip2 in partits:
        gols_fets[equip1] += gol1
        gols_fets[equip2] += gol2
        gols_rebuts[equip1] += gol2
        gols_rebuts[equip2] += gol1

    return gols_fets, gols_rebuts

def calcular_punts(partits):
    """Calcula els punts de cada equip segons el resultat dels partits."""
    punts = defaultdict(int)

    for equip1, gol1, gol2, equip2 in partits:
        if gol1 > gol2:
            punts[equip1] += 3
        elif gol2 > gol1:
            punts[equip2] += 3
        else:
            punts[equip1] += 1
            punts[equip2] += 1

    return punts

def equip_mes_golejador(gols_fets):
    return max(gols_fets, key=lambda k: gols_fets[k])

def equip_mes_golejat(gols_rebuts):
    return max(gols_rebuts, key=lambda k: gols_rebuts[k])

def classificacio_final(punts):
    return sorted(punts.items(), key=lambda x: x[1], reverse=True)

def mostrar_resultats(gols_fets, gols_rebuts, punts):
    golejador = equip_mes_golejador(gols_fets)
    golejat = equip_mes_golejat(gols_rebuts)
    classificacio = classificacio_final(punts)

    print("Gols per equip:")
    for equip, gols in gols_fets.items():
        print(f"{equip}: {gols} gols")

    print(f"\nEquip més golejador: {golejador} ({gols_fets[golejador]} gols)")
    print(f"Equip més golejat: {golejat} ({gols_rebuts[golejat]} gols rebuts)")

    print("\nClassificació global:")
    for i, (equip, pts) in enumerate(classificacio, start=1):
        print(f"{i}. {equip}: {pts} pts")

def principal_partits():
    partits = llegir_partits("historic partits.txt")
    gols_fets, gols_rebuts = calcular_gols(partits)
    punts = calcular_punts(partits)
    mostrar_resultats(gols_fets, gols_rebuts, punts)

principal_partits()
```
